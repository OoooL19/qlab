```python
import pandas as pd
from pandas import DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)

df = pd.read_excel('/Users/julieschneider/Julie_Personal/Projects/HEB/HEB_data.xlsx', sheet_name='Sheet1')
df.rename(columns={'Linguistic SL':'Percent Correct'}, inplace=True)
df.rename(columns={'SL':'Statistical Learning'}, inplace=True)

print("Column headings:")
print(df.columns)
print(df)
```

```python
#sns.load_dataset("HEB_data")
sns.catplot(x="Statistical Learning", y="Percent Correct", hue="Language", kind="bar", data=df);
```
